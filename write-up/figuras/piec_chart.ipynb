```python
import matplotlib.pyplot as plt
# plt dpi
plt.rcParams['figure.dpi'] = 200
# size of the labels
plt.rcParams['font.size'] = 20

# Data for the pie chart
labels = ['Entrenamiento', 'Validacin', 'Evaluacin']
sizes = [3139, 393, 392]

# Create the pie chart with actual numbers displayed
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=[f"{label}: {size}" for label, size in zip(labels, sizes)], 
        startangle=60)
# position the legend
plt.show()
```

```python
# Create a pie chart focusing on the "Evaluacin" slice with the others faded out

# Adjust the explosion to highlight the "Evaluacin" slice
explode = (0, 0, 0.2)  # Only explode the third slice

# Create the pie chart with a highlighted "Evaluacin" slice
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, explode=explode, autopct=lambda p: f'{int(p * sum(sizes) / 100)}',
        startangle=90)
plt.show()
```
